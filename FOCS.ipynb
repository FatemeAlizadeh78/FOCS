```python
PATH = "ZTBus_compressed/"
```

```python
def merge_files_in_df(path: str):
    
    files = os.listdir(path)
    output_df = pd.DataFrame()
    c = 0
    d = 0
    for file in files:
        if ('B183' in file) and (c < 300):
            df = pd.read_csv(f'{path}{file}')
            df["busNumber"] = file[1:4] 
            df["missionDate"] = file[5:15]
            df["startMissionTime"] = file[15] 
        
            print(file)
            output_df = pd.concat([output_df, df])
            c = c + 1
        if ('B208' in file) and (d < 300):
            df = pd.read_csv(f'{path}{file}')
            df["busNumber"] = file[1:4] 
            df["missionDate"] = file[5:15]
            df["startMissionTime"] = file[15] 
        
            print(file)
            output_df = pd.concat([output_df, df])
            d = d + 1
    return output_df
```

## 9. For each (route, bus) pair, compute the ratio between the overall energy consumption and the overall driven distance.


```python
bus_route_group = trips_df.groupby(["busNumber", "busRoute"])
energy_distance_ratio = (bus_route_group["energyConsumption"].sum() / bus_route_group["drivenDistance"].sum()).reset_index(name="ratio")
energy_distance_ratio
```

## 10. Starting from the results of the previous point, for each route compute the buses with max and min energy ratio, and save the difference between these ratios in a dataframe.


```python
max_energy_ids = energy_distance_ratio.groupby("busRoute")["ratio"].idxmax()
min_energy_ids = energy_distance_ratio.groupby("busRoute")["ratio"].idxmin()
bus_max_consumption = energy_distance_ratio.loc[max_energy_ids].reset_index()

bus_min_consumption = energy_distance_ratio.loc[min_energy_ids].reset_index()

ratio = bus_max_consumption["ratio"] - bus_min_consumption["ratio"]
energy_distance_ratio["ratio_interval_diff"] = ratio
energy_distance_ratio
```

```python
max_energy_ids, min_energy_ids
```

## 11. Find the bus maximizing the difference computed in the previous point.


```python
max_id = energy_distance_ratio["ratio_interval_diff"].idxmax()
energy_distance_ratio.loc[max_id]
```

## 15. For each bus compute the total time when the halt brake is active and the total time when the park brake is active. Compute also the ratio between those two times.


### Halt brake

```python
details = details.sort_values(by=["busNumber", "missionDate", "time_unix"])
details["diff_second"] = details.groupby(["busNumber", "missionDate"])["time_unix"].diff()
details["diff_second"] = details["diff_second"].fillna(0)

halt_brake_time = details[details["status_haltBrakeIsActive"] == 1]
halt_brake_time_df = halt_brake_time.groupby(["busNumber"])["diff_second"].sum()
```

```python
halt_brake_time_df
```

### Park brake

```python
details = details.sort_values(by=["busNumber", "missionDate", "time_unix"])
details["diff_second_2"] = details.groupby(["busNumber", "missionDate"])["time_unix"].diff()
details["diff_second_2"] = details["diff_second_2"].fillna(0)

park_brake_time = details[details["status_parkBrakeIsActive"] == 1]
park_brake_time_df = park_brake_time.groupby(["busNumber"])["diff_second_2"].sum()
```

```python
park_brake_time_df
```

```python
halt_brake_time_df / park_brake_time_df
```
